#### Training SVD, KNNBasic & KNNWithMeans Models

To ensure we select the best-performing recommendation model for our system, we trained and evaluated three popular `collaborative filtering algorithms`: `SVD`, `KNNBasic`, and `KNNWithMeans`. These models were chosen for their distinct approaches to predicting user ratings. `Singular Value Decomposition (SVD)` is known for its ability to handle sparse data and uncover latent factors in user-item interactions. KNNBasic and KNNWithMeans use similarity-based approaches to make recommendations, with KNNWithMeans incorporating mean-centered ratings to enhance prediction accuracy.

The evaluation was conducted using cross-validation to obtain reliable performance metrics, specifically focusing on the Root `Mean Square Error (RMSE)` to measure prediction accuracy. 

```python

def evaluate_models(data):
    # Split the data into training and test sets
    trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

    # Define a list of models to evaluate
    models = [
        SVD(),
        KNNBasic(sim_options={'name': 'pearson', 'user_based': True}),
        KNNWithMeans(sim_options={'name': 'pearson', 'user_based': True})
    ]

    # Evaluate each model and store the results
    results = []
    for model in models:
        # Perform cross-validation
        cv_results = cross_validate(model, data, measures=['RMSE'], cv=5, verbose=False)

        # Get the average RMSE from cross-validation
        rmse = cv_results['test_rmse'].mean()

        # Store the model and its performance
        results.append({'model': model.__class__.__name__, 'rmse': rmse})

    # Convert the results to a DataFrame
    performance_df = pd.DataFrame(results)
    
    return performance_df
evaluate_models(data)
```

#### Model Comparison with RMSE

```python
performance_df = evaluate_models(data)

# Plot the performance of each model
plt.figure(figsize=(12, 8))
sns.barplot(x='model', y='rmse', data=performance_df, palette='viridis')
plt.title('Model Performance Comparison', fontsize=20, weight='bold')
plt.xlabel('Model', fontsize=16, weight='bold')
plt.ylabel('RMSE', fontsize=16, weight='bold')
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

These results indicate that:

* `SVD` (Singular Value Decomposition) achieved the lowest RMSE of `0.874106`, making it the best performing model among the three. This suggests that SVD is the most accurate model for predicting user ratings in our dataset.

* `KNNBasic` (K-Nearest Neighbors Basic) had the highest RMSE of `0.973791`, indicating that its predictions are less accurate compared to the other models. This could be due to its simpler approach, which does not take into account mean rating adjustments.

* `KNNWithMeans` (K-Nearest Neighbors with Means) had an RMSE of `0.896503`, performing better than KNNBasic but not as well as SVD. This model improves upon KNNBasic by adjusting for user rating biases, resulting in more accurate predictions.

 
The SVD model is the preferred choice for our recommendation system due to its superior performance in minimizing prediction errors.

#### Fit SVD Model

After evaluating multiple models, we identified SVD as the best performing model. To optimize its performance, we conducted a Grid Search to identify the best hyperparameters. The following steps outline the fitting of the SVD model with the optimal parameters.

```python
# Define the parameter grid
params = {'n_factors': [20, 50, 100], 'reg_all': [0.02, 0.05, 0.1]}

# Perform Grid Search
g_s_svd = GridSearchCV(SVD, param_grid=params, n_jobs=-1, measures=['rmse'], cv=5)
g_s_svd.fit(data)

# Output the best score and best parameters
print("Best RMSE Score: ", g_s_svd.best_score['rmse'])
print("Best Parameters: ", g_s_svd.best_params['rmse'])
```

The Grid Search identified `n_factors = 50` and `reg_all = 0.05` as the optimal hyperparameters for the SVD model, resulting in an `RMSE` of `0.8689835790309699`. This indicates a slight improvement over the initial evaluation, confirming the SVD model's effectiveness in predicting user ratings accurately.

By fitting the `SVD` model with these parameters, we ensure that our recommendation system is well-tuned and ready for deployment. The `SVD` model can now predict movie ratings with improved accuracy, which will enhance the user experience on the streaming platform.

```python
# Split the data into training and test sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
#Train SVD Model 
svd = SVD(n_factors=100, reg_all=0.05)
svd.fit(trainset)
```

#### Collect Movie Ratings from the User

To enhance the recommendation system's accuracy, it is essential to collect explicit feedback from users. The following function allows users to rate movies, which helps in personalizing recommendations based on their preferences. This process involves asking the user to rate a set number of movies, with an option to filter by their preferred genre.

```python
# Function to collect movie ratings from the user
def get_movie_ratings(movie_df, num_ratings, preferred_genre=None):
    user_id = 1000
    ratings = []
    missed_count = 0  
    
    while num_ratings > 0:
        # Filter movies based on genre if provided
        if preferred_genre:
            movie_sample = movie_df[movie_df['genres'].str.contains(preferred_genre, case=False, na=False)].sample(1)
        else:
            movie_sample = movie_df.sample(1)
        
        movie_title = movie_sample['title'].values[0]
        movie_id = movie_sample['movieId'].values[0]
        
        print(f"\nMovie: {movie_title}")
        rating_input = input('Rate this movie from 1 to 5, or press "n" if you have not seen it:\n')
        
        if rating_input.lower() == 'n':
            missed_count += 1
            if missed_count >= 3 and not preferred_genre:
                preferred_genre = input('You have skipped several movies. Could you please tell us your favorite genre?\n')
            continue
        elif rating_input.isdigit() and 1 <= int(rating_input) <= 5:
            ratings.append({'userId': user_id, 'movieId': movie_id, 'rating': float(rating_input)})
            num_ratings -= 1
            missed_count = 0  
        else:
            print('Invalid input. Please enter a number between 1 and 5 or "n".')
    
    return ratings


df_movies = pd.read_csv('Data/movies.csv')  
user_ratings = get_movie_ratings(df_movies, 4, 'Comedy')

# Print collected ratings
print("\nCollected Ratings:")
for rating in user_ratings:
    print(f"User ID: {rating['userId']}, Movie ID: {rating['movieId']}, Rating: {rating['rating']}")
```

The function successfully collects user ratings, which are essential for tailoring movie recommendations. These ratings will be incorporated into the recommendation system, enabling it to generate more accurate and personalized movie suggestions based on the user's preferences.

```python
# Convert the list of new ratings to a DataFrame
user_ratings_df = pd.DataFrame(user_ratings)

# Load the original ratings data
original_ratings_df = pd.read_csv('Data/ratings.csv')

# Combine the original ratings with new user ratings
combined_ratings_df = pd.concat([original_ratings_df, user_ratings_df], axis=0)


combined_ratings_df = combined_ratings_df[['userId', 'movieId', 'rating']]
combined_ratings_df['rating'] = combined_ratings_df['rating'].astype(float)

# Define the rating scale for Surprise
reader = Reader(rating_scale=(0.5, 5.0))

# Create a new Dataset object from the updated DataFrame
new_data = Dataset.load_from_df(combined_ratings_df, reader)
```

#### Predictions

To provide personalized movie recommendations for the user with `userId = 1000`, we used the trained SVD model to predict ratings for all movies in the dataset. The predictions were then sorted from highest to lowest to identify the top recommendations.

```python
# Generate movie recommendations for the user with userId = 1000
list_of_movies = []
for movie_id in ratings_df['movieId'].unique():
    # Predict the rating the user would give to each movie
    predicted_rating = svd.predict(1000, movie_id).est
    list_of_movies.append((movie_id, predicted_rating))

# Order the recommendations from highest to lowest rated
ranked_movies = sorted(list_of_movies, key=lambda x: x[1], reverse=True)

# Convert the ranked movies to a DataFrame for better readability
recommendations_df = pd.DataFrame(ranked_movies, columns=['movieId', 'predicted_rating'])

# Display the top 10 recommendations
top_10_ranked = recommendations_df.head(10)
print(top_10_ranked)
```

The top 5 movie recommendations for the user, generated by the SVD model, highlight films with the highest predicted ratings based on their preferences and previous ratings. To ensure the model trained for the recommendation system is available for future use without retraining, we employ model persistence techniques. The model is saved to disk using a serialization method, allowing it to be reused or deployed in different environments. When needed, the saved model can be loaded from disk to make predictions, ensuring that the deployment process is efficient and that the latest trained model is utilized.
